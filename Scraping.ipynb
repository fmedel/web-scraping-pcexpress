```python
f = open('./data.csv','r')
data = pd.read_csv(f,sep=",")
```

```python
stack_a=[]
stack_b=[]
stack_c=[]
stack_d =[]
for x in range(0,300):
    url='https://tienda.pc-express.cl/index.php?route=product/category&path='+str(x)+'&limit=100'
    response = requests.get(url)
    html = BeautifulSoup(response.text, "html.parser")
    nombre_catagoria = html.find('div',{'id':'content'})
    try:
        if(nombre_catagoria.h1.text=="Categories"):
            print(x,"no funca")
        else:
            print(x,"funca")
    except:
        entradas = html.find_all('div', {'class':'product-list__item'})
        for datos in entradas:
            nombre=datos.find('h5', {'class':'product-list__name'})
            marca=datos.find('p', {'class':'pull-left'})
            valor=datos.find('p', {'class':'product-list__price'})
            stack_a.append(nombre.a.text)
            stack_b.append(marca.text)
            stack_c.append(valor.text)
            stack_d.append(nombre_catagoria.h2.text)

```

```python
data.drop('nombre', axis=1, inplace=True)
data.drop(' marca', axis=1, inplace=True)
data.drop('valor', axis=1, inplace=True)
data.drop('categoria', axis=1, inplace=True)
```

```python
data['nombre']=stack_a
data['marca']=stack_b
data['valor'] = stack_c
data['categoria']=stack_d
```

```python
data.to_csv("data_bi/todos.csv")
```
